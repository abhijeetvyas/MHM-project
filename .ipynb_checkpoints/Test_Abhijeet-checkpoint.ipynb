```python
import scipy.fftpack
```

```python
N=106
w = scipy.fftpack.rfft(ankle_y)
f = scipy.fftpack.rfftfreq(N, time[1]-time[0])
spectrum = w**2

cutoff_idx = spectrum < (spectrum.max()/4000)
w2 = w.copy()
w2[cutoff_idx] = 0

y2 = scipy.fftpack.irfft(w2)
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.plot(time,y2)
```

```python
from scipy.signal import filtfilt
```

```python
n = 3# the larger n is, the smoother curve will be 
b = [1.0 / n] * n 
a = 1 
yy = filtfilt(b,a,ankle_y) 
new=plt.plot(time, yy, linewidth=2, linestyle="-", c="b")
```
