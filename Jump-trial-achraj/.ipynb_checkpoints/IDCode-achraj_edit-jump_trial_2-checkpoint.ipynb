from the filtered position data, find the segment angles

```python
theta_f=dy.absangle(Txp,Typ,Axp,Ayp,106)
theta_f = dy.filterdata(time,theta_f,5)
```

```python
theta_s = dy.absangle(Axp,Ayp,Kxp,Kyp,106)
theta_s = dy.filterdata(time,theta_s,5)
```

```python
alpha_f = dy.derivative(omega_f,time)
alpha_f = dy.filterdata(time,alpha_f,5)
```

```python
alpha_s = dy.derivative(omega_s,time)
alpha_s = dy.filterdata(time,alpha_s,4)
```

```python
alpha_t = dy.derivative(omega_t,time)
alpha_t = dy.filterdata(time,alpha_t,3)
```

The Inverse dynamics analysis is done by segmenting the body into foot, shank, thigh and the upper body is considered to be a lumped mass at the centre of mass of the upper body

*add the fbd of the whole body*

consider the foot

*add the fbd of the foot*

The kinetic diagram of the foot is as shown

*add kinetic diagram of the foot*

*add the equations here*

```python
Ax,Ay,MA=dy.IDfoot(Rx,Ry,theta_f,alpha_f,mf,If,lf,lcf,afx,afy)
```

Consider the shank

*add FBD of shank*

*add kinetic diagram*

```python
Kx,Ky,Mk=dy.IDlimb(Ax,Ay,MA,theta_s,alpha_s,ms,Is,ls,lcs,asx,asy)
```

consider the thigh

*add FBD of thigh*

*add kinetic diagram*

```python
Hx,Hy,Mh=dy.IDlimb(Kx,Ky,Mk,theta_t,alpha_t,mt,It,lt,lct,atx,aty)
```

```python
#filter the Joint forces and moments
#find the joint angles and joint omegas.
```

```python
theta_A=theta_f-theta_s
theta_K=theta_s-theta_t
theta_H=-np.pi/2+theta_t
```

```python
omega_A=omega_f-omega_s
omega_K=omega_s-omega_t
omega_H=omega_t
```

```python
Power_A=MA*omega_A
Power_K=Mk*omega_K
power_H=Mh*omega_H
```

```python
Power_A=dy.filterdata(time,Power_A,3)
```

```python
Power_K=dy.filterdata(time,Power_K,3)
```
