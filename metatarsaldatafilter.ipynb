```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import filtfilt
```

```python
df=pd.read_csv('F:\Python Workspace\DOCUMENTS\Raw_gait_data_N1 set.csv', sep=',')
```

```python
t=df.loc[:,"TIME"]
x=df.loc[:,"Right metatarsal X"]
y=df.loc[:,"Right metatarsal Y"]
```

```python
k=df.plot(x="TIME",y="Right metatarsal X")
```

```python
n = 10 # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1
xx = filtfilt(b,a,x)
plt.plot(t, xx, linewidth=2, linestyle="-", c="b")
```

```python
k=df.plot(x="TIME",y="Right metatarsal Y")
```

```python
n = 6 # the larger n is, the smoother curve will be
b = [1.0 / n] * n
a = 1
yy = filtfilt(b,a,y)
plt.plot(t, yy, linewidth=2, linestyle="-", c="b")
```
